# Data Cleaning, Analysis, and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('dataset.csv')
df.head()
```

```python
# Clean and process the dataset
df_clean = df.drop(columns=['Unnamed: 0', 'salary', 'salary_currency'])
df_clean.rename(columns={
    'work_year': 'Year',
    'experience_level': 'Experience',
    'employment_type': 'EmploymentType',
    'job_title': 'JobTitle',
    'salary_in_usd': 'SalaryUSD',
    'employee_residence': 'Residence',
    'remote_ratio': 'RemoteRatio',
    'company_location': 'CompanyLocation',
    'company_size': 'CompanySize'
}, inplace=True)

experience_map = {'EN': 'Entry-level', 'MI': 'Mid-level', 'SE': 'Senior', 'EX': 'Executive'}
employment_map = {'FT': 'Full-time', 'PT': 'Part-time', 'CT': 'Contract', 'FL': 'Freelance'}
company_size_map = {'S': 'Small', 'M': 'Medium', 'L': 'Large'}

df_clean['Experience'] = df_clean['Experience'].map(experience_map)
df_clean['EmploymentType'] = df_clean['EmploymentType'].map(employment_map)
df_clean['CompanySize'] = df_clean['CompanySize'].map(company_size_map)
df_clean.head()
```

```python
# Visualize average salary by experience level
plt.figure(figsize=(10, 6))
sns.barplot(data=df_clean, x='Experience', y='SalaryUSD', estimator=pd.Series.mean, ci=None)
plt.title('Average Salary by Experience Level')
plt.ylabel('Average Salary (USD)')
plt.xlabel('Experience Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Additional Analyses and Visualizations

```python
# Average Salary by Job Title (Top 10 by Frequency)
top_jobs = df_clean['JobTitle'].value_counts().head(10).index
plt.figure(figsize=(12, 6))
sns.barplot(data=df_clean[df_clean['JobTitle'].isin(top_jobs)],
            x='JobTitle', y='SalaryUSD', estimator=pd.Series.mean, ci=None)
plt.title('Average Salary by Top 10 Job Titles')
plt.ylabel('Average Salary (USD)')
plt.xlabel('Job Title')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Salary Distribution by Company Size
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_clean, x='CompanySize', y='SalaryUSD')
plt.title('Salary Distribution by Company Size')
plt.ylabel('Salary (USD)')
plt.xlabel('Company Size')
plt.tight_layout()
plt.show()
```

```python
# Salary vs Remote Ratio Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='RemoteRatio', y='SalaryUSD', hue='Experience')
plt.title('Salary vs Remote Work Ratio')
plt.ylabel('Salary (USD)')
plt.xlabel('Remote Work Ratio (%)')
plt.tight_layout()
plt.show()
```
